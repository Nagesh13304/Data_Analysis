# World Population Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('population.xlsx')
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
#Changing the Data type of LocID into int
df['LocID']=df['LocID'].astype(int)
df.info()
```

```python
#Changing the Data type of Time col into int
df['Time']=df['Time'].astype(int)
df.info()
```

```python
#Changing the Data type of AgeGrpStart col into int
df['AgeGrpStart']=df['AgeGrpStart'].astype(int)
df.info()
```

```python
#Removing unwanted data in PopMale 'ERROR_6.246'

df['PopMale']=df['PopMale'].str.strip('ERROR_')

#Changing the Data type of popmale PopFemale col into int
df['PopMale']=df['PopMale'].astype(float)
df['PopMale']=df['PopMale'].astype(int)
df['PopFemale']=df['PopFemale'].astype(int)
df.info()
```

**Visualizing Total Population By Location**

```python
loc=df.groupby('Location')['PopTotal'].sum().sort_values(ascending=False)
loc=pd.DataFrame(loc).iloc[:7,:]
plt.figure(figsize=(12,5))
a=sns.barplot(x='Location',y='PopTotal',data=loc,hue='Location',palette='plasma',errorbar=None)
a.set_xlabel('Location',fontsize=13,color='green')
a.set_ylabel('Total Population',fontsize=13,color='green')
def data_label(a):
  for i in a.containers:
    a.bar_label(i)

data_label(a)
```

Total population of India and China is more compared to other locations

# Visualizing Population male and female by Location

```python
pop=df.pivot_table(values=['PopMale','PopFemale'],index='Location',aggfunc='sum')
pop.sort_values(by='PopMale',ascending=False,inplace=True)
pop=pop.iloc[:7,:]

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,5))
a=sns.barplot(x='Location',y='PopMale',hue='Location',data=pop,palette='plasma',ax=axes[0])

b=sns.barplot(x='Location',y='PopFemale',hue='Location',data=pop,palette='plasma',ax=axes[1])
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=90)
axes[1].set_xticklabels(axes[1].get_xticklabels(),rotation=90)
data_label(a)
data_label(b)
plt.show()
```

Male and Female population is more in India and China compared to other locations

**Visualizing Total Population by AgeGrp**

```python
age=df.groupby(['AgeGrp'])['PopTotal'].sum().sort_values(ascending=False)
age=pd.DataFrame(age).iloc[:7,:]
# plt.figure(figsize=(17,5))
# a=sns.barplot(x='AgeGrp',y='PopTotal',hue='AgeGrp',data=age,palette='plasma')
# for i in a.containers:
#   a.bar_label(i,label_type='center',rotation=90)

plt.pie(age['PopTotal'],labels=['0-4','5-9','10-14','15-19','20-24','25-29','30-34'],autopct='%0.0f%%')
plt.show()
```

Population Under Age group 0-4,5-9,10-14, 15-19 is more compared to other age groups

```python
time=df.groupby('Time')['PopTotal'].sum()
time=pd.DataFrame(time).reset_index()
plt.figure(figsize=(17,5))
a=sns.lineplot(x='Time',y='PopTotal',data=time,marker='o',markersize=8,markerfacecolor='orange',markeredgecolor='orange')
for i,j in zip(time['Time'],time['PopTotal']):
  plt.annotate(f'{i},{j}',(i,j),ha='left')
```

Total Population over Time is Increasing
