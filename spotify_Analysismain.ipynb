```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify_data clean.csv",encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
#handling artist_genres column
df["artist_genres"] = df["artist_genres"].fillna("Unknown")
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
#to remove  --> track_id, explicit, album_id
df = df.drop(["track_id", "explicit", "album_id"],axis=1)
df.head()
```

```python
df["album_release_date"] = pd.to_datetime(df["album_release_date"])
df.info()
```

```python
df.head()
```

```python
df["release_year"] = df["album_release_date"].dt.year
df.head()
```

```python
# count of albums released over year
color_code = "#1db954"
plt.hist(df["release_year"],color=color_code)
plt.xlabel("Release Year")
plt.ylabel("Count of Albums")
plt.title("Count of Albums over year")
plt.show()
```

```python
#count by album_type

plt.hist(df["album_type"],color=color_code)
plt.xlabel("Album type")
plt.ylabel("Count of Album type")
plt.title("Count by Album type")
plt.show()
```

```python
# Avg track popularity by album type
data = df.groupby("album_type")["track_popularity"].mean()
data = pd.DataFrame(data).reset_index()
plt.plot(data["album_type"],data["track_popularity"],color=color_code,marker="o",markerfacecolor="#212121",markeredgecolor="#212121")
plt.xlabel("Album Type")
plt.ylabel("Avg Track popularity")
plt.title("Avg track popularity by album type")
plt.show()
```
